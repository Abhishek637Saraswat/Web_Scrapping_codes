```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re



# Define the User-Agent headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
}


count = 0
for link in links:
    url = []
    for i in range(1, 20):
        k = i * 50
        url1 = str(link) + str(k)
        url.append(url1)

    count += 1
    print(f'Count: {count}')
    print(f'URLs generated: {len(url)}')

    for u in url:
        page = requests.get(u, headers=headers)
        soup = BeautifulSoup(page.text, 'html.parser')
        href_elements = soup.find_all('a', class_='base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]')
        urls = [element.get('href') for element in href_elements]
        print(len(urls))

        for job_url in urls:
            try:
                job_page = requests.get(job_url, timeout=10)
                job_page.raise_for_status()  # Raises HTTPError for bad responses (4xx and 5xx)

            except requests.RequestException as e:
                 print(f"Failed to retrieve {job_url}: {e}")
                 continue

            
            job_soup = BeautifulSoup(job_page.text, 'html.parser')

            # Extract company name
            company_element = job_soup.find('a', class_='topcard__org-name-link topcard__flavor--black-link')
            company_name = company_element.get_text(strip=True) if company_element else 'N/A'

            # Extract job description
            description_element = job_soup.find('div', class_='show-more-less-html__markup show-more-less-html__markup--clamp-after-5 relative overflow-hidden')
            html_tag_pattern = re.compile(r'<.*?>')
            cleaned_content = re.sub(html_tag_pattern, '\n', str(description_element))

            # Extract job title
            title_element = job_soup.find('h1', class_='top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title')
            job_title = title_element.get_text(strip=True) if title_element else 'N/A'

            # Create a DataFrame with the new row
            new_row = pd.DataFrame({
                'job_title': [job_title],
                'company_name': [company_name],
                'description': [cleaned_content]
            })

            # Append the new row to the existing DataFrame using pd.concat
            df5 = pd.concat([df5, new_row], ignore_index=True)

# Display the final DataFrame
df5
```
